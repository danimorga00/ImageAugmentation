```python
import pandas as pd
import matplotlib.pyplot as plt

# Carica il CSV senza header
df = pd.read_csv('results1', header=None)

# Scegli le colonne da plottare
colonna_x = 0  # Indice della colonna da usare come asse x
colonna_y1 = 1  # Indice della prima colonna da usare come asse y

# Crea il plot
plt.figure(figsize=(10, 6))

prima_valore = df[1].iloc[0] 
# Calcola il rapporto tra le due colonne
df['rapporto'] = prima_valore / df[1]
    
# Plotta la prima colonna in funzione di colonna_x
plt.plot(df[colonna_x], df[colonna_y1]/100, marker='o', label='Time (seconds/100)')
# Plotta il rapporto
plt.plot(df[colonna_x], df['rapporto'], marker='x', label='Speed up', linestyle='--')


# Aggiungi titoli e etichette
plt.title('Versione parallela')
plt.xlabel(f'workers')
plt.legend()  # Mostra la legenda
plt.grid()
plt.show()
```
